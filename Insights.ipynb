```python
models = {
    'Model_A': X_scaled.columns,

    'Model_B': X_scaled.drop(columns='bathrooms').columns,

    'Model_C': X_scaled.drop(columns='bhk').columns
}
```

```python
results = []

for name, features in models.items():

    X_subset = X_scaled[features]

    scores = cross_val_score(
        LinearRegression(),
        X_subset,
        y_log,
        cv=5,
        scoring='r2'
    )

    results.append({
        'Model': name,
        'Features': len(features),
        'Mean CV R²': scores.mean(),
        'Std CV R²': scores.std()
    })
```

```python
results_df = pd.DataFrame(results)

print(results_df.sort_values('Mean CV R²', ascending=False))
```

```python
X_scaled_v2 = X_scaled.drop(columns='bathrooms')
```

```python
lr.fit(X_scaled_v2,y_log)
```

```python
coef_df_v2 = pd.DataFrame(lr.coef_.reshape(1,15),columns=X_scaled_v2.columns).stack().reset_index().drop(columns=['level_0']).rename(columns={'level_1':'feature',0:'coef'})
coef_df_v2
```

```python
feature_stds = X[X_scaled_v2.columns].std()

unscaled_coefficients = (
    coef_df_v2.set_index('feature')['coef']
    / feature_stds
)
display(unscaled_coefficients)
```

```python
feature_stds = X[X_scaled_v2.columns].std()

beta = (
    coef_df_v2
    .set_index('feature')['coef']
    / feature_stds
)

results = []

for feature, b in beta.items():

    # Area: interpret as +100 sqft
    if feature == 'buildup_area_sqft':

        effect_pct = (np.exp(b * 100) - 1) * 100

        interpretation = "+100 sqft"

    # Dummy variables: interpret as 0 -> 1
    elif (
        feature.startswith('location_zone_')
        or feature.startswith('age_category_')
        or feature == 'is_ready_to_move'
    ):

        effect_pct = (np.exp(b) - 1) * 100

        interpretation = "0 -> 1"

    else:

        effect_pct = (np.exp(b) - 1) * 100

        interpretation = "+1 unit"

    results.append({
        'feature': feature,
        'beta': round(b, 6),
        'interpretation': interpretation,
        'price_effect_%': round(effect_pct, 2)
    })

inference_df = pd.DataFrame(results)

display(
    inference_df.sort_values(
        'price_effect_%',
        ascending=False
    )
)
```

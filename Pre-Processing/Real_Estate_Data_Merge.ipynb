```python
def isL(text):

  cleaned = text.replace("Rs. ","").strip().replace(" Cr","").split(" - ")

  if len(cleaned) == 1:
    cleaned.append(cleaned[0])

  if "L" in text:

    if "L" in cleaned[0]:
      cleaned[0] = cleaned[0].replace("L", "").strip()
      cleaned[0] = round(float(cleaned[0])/100,2)

    if "L" in cleaned[1]:
      # Fixed: update cleaned[1] here, not [0] again
      cleaned[1] = cleaned[1].replace("L", "").strip()
      cleaned[1] = round(float(cleaned[1])/100,2)

  # Minimal change: Ensure values are explicitly cast to float
  return [float(cleaned[0]), float(cleaned[1])]

df1[['min_price', 'max_price']] = df1['pricedetails'].apply(isL).tolist()
```

```python
df1.head()
```

```python
new_data = []

for i in range(len(df1)):
  text = df1.loc[i, 'propertysubname']
  bhk_list = text.replace(" BHK Flat","").strip().split(", ")

  prices = list(np.linspace(
          df1.loc[i,'min_price'],
          df1.loc[i,'max_price'],
          len(bhk_list)
      ).round(2))

  for bhk, price in zip(bhk_list, prices):
    new_row = df1.loc[i].copy()
    new_row['propertysubname'] = str(bhk) + " BHK Flat"
    new_row['pricedetails'] = round(price, 2)
    new_data.append(new_row)

new_df = pd.DataFrame(new_data)
new_df
```

```python
new_df.drop(columns=['bhk_list',	'min_price',	'max_price'], inplace=True)
```

```python
df2['source'] = 'kaggle'
```

```python
df2['location'] = df2['address'].str.split(', ').str.get(-2)
```

```python
def convert_price(text):

    text = str(text).strip()

    if "Cr" in text:
        return float(text.replace("Cr", "").strip())

    elif "L" in text:
        return round(
            float(text.replace("L", "").strip()) / 100,
            2
        )

    return float(text)

df2['price'] = df2['price'].apply(convert_price)
```

```python
# 1. Filter the dataset to only include rows that contain 'BHK Flat'
df_flats = df2[df2['flat_type'].str.contains('BHK Flat', na=False)].copy()

# 2. Extract the leading digits, convert them to integer
df_flats['bhk_type'] = df_flats['flat_type'].str.extract(r'(\d+)').astype(int)

# 3. Ensure 'bedrooms' is numeric for a correct comparison
df_flats['bedrooms'] = pd.to_numeric(df_flats['bedrooms'], errors='coerce')

# 4. Find and print rows where flat_type number does not match bedrooms column
mismatch = df_flats[df_flats['bedrooms'] != df_flats['bhk_type']]

print(f"Total Mismatches found: {len(mismatch)}")
print(mismatch[['flat_type', 'bhk_type', 'bedrooms']].head(10))
```

```python
# 1.5 BHK will break if you cast bhk_type to int
print(df2['flat_type'].unique())  # check for 1.5, 2.5, 3.5 BHK etc.
```

```python
df2 = df2[df2['flat_type'].str.contains('Flat', na=False)].reset_index(drop=True)
```

```python
def parse_sqft(val):
    if pd.isna(val):
        return np.nan

    val_str = str(val).replace('sq.ft', '').replace(',', '').strip()

    if '-' in val_str:
        parts = val_str.split('-')
        try:
            return (float(parts[0].strip()) + float(parts[1].strip())) / 2
        except ValueError:
            return np.nan

    try:
        val_float = float(val_str)
        return val_float if val_float >= 100 else np.nan
    except ValueError:
        return np.nan

df2['buildup_area_sqft'] = df2['buildup_area'].apply(parse_sqft)
df2['carpet_area_sqft']  = df2['carpet_area'].apply(parse_sqft)
```

```python
df2['furnishing'] = df2['facing'].apply(
    lambda x: x if pd.notna(x) and 'urnished' in str(x) else np.nan
)
df2['furnishing'] = df2['furnishing'].fillna('Unknown')
```

```python
df2['is_ready_to_move'] = df2['possesion_status'] == 'Ready to move'
```

```python
def parse_age(val):
    val = str(val).strip()
    if val == 'Ready to move':
        return 0
    # BHK values that got misplaced — treat as NaN
    if 'BHK' in val:
        return np.nan
    first = val.split(' ')[0]
    try:
        return int(first)
    except ValueError:
        return np.nan

df2['age_of_property'] = df2['age_of_property'].apply(parse_age)
```

```python
def parse_floor(val):
    if pd.isna(val): return np.nan
    val = str(val)
    if 'of' in val:
        position = val.split('of')[0].strip()
        if position in ('Higher', 'Middle', 'Lower'):
            return position
    return np.nan

df2['which_floor'] = df2['which_floor'].apply(parse_floor)
```

```python
def parkingCount(text):
    if pd.isna(text) or str(text).strip() == '-':
        return 0

    text = str(text)

    if 'Covered' in text and 'Open' in text:
        num1 = int(text.split(' ')[0])
        num2 = int(text.split(' ')[-3])
        return num1 + num2

    elif 'Covered' in text or 'Open' in text:
        return int(text.split(' ')[0])

    return 0

df2['parking'] = df2['parking'].apply(parkingCount)
```

```python
# fix bedrooms: drop misplaced values
df2['bedrooms'] = pd.to_numeric(df2['bedrooms'], errors='coerce')
```

```python
# drop columns no longer needed
df2.drop(columns=[
    'address', 'buildup_area', 'carpet_area',
    'avg_price_sqft', 'facing', 'furnicing', 'brokrage',
    'about_property'
], inplace=True)
```

```python
mask = df2['bathrooms'] == 'No Balcony'
df2.loc[mask, 'balcony'] = 'No Balcony'
df2.loc[mask, 'bathrooms'] = np.nan
```

```python
mask = ~df2['possesion_status'].str.contains("Ready to move", na=False)
df2.loc[mask,'possesion_status'] = np.nan
```

```python
df2 = df2[['flat_type', 'price', 'age_of_property', 'possesion_status', 'which_floor', 'project_name', 'bedrooms', 'bathrooms', 'balcony', 'parking', 'source', 'location', 'buildup_area_sqft', 'carpet_area_sqft', 'is_ready_to_move', 'furnishing']]
```

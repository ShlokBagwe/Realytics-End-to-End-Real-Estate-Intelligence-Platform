## Price

```python
(df['price'].str.contains('request', case=False, na=False)).sum()
```

```python
df= df[~df['price'].str.contains('request', case=False, na=False)].reset_index(drop=True)
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
missing_unit = df[~df['price'].str.contains('Lac|Crore', case=False, na=False)]

if missing_unit.empty:
    print("All entries contain either 'Lac' or 'Crore'.")
else:
    print(f"Found {len(missing_unit)} entries without standard units:")
    display(missing_unit[['price']])
```

```python
def priceCleaning(text):

    text = text.replace("Rs. ", "").strip()

    if "Crore" in text:
        text = text.replace(" Crore", "").strip()
        return round(float(text),2)

    if "Lac" in text:
        text = text.replace(" Lac", "").strip()
        return round(float(text) / 100,2)


df['price'] = df['price'].apply(priceCleaning)
```

## features

```python
df['features'].value_counts()
```

```python
def clean_features(text):
    if pd.isna(text):
        return text
    parts = [p.strip() for p in text.split('|')]
    parts = [p for p in parts if not p.startswith('#')]
    return ' | '.join(parts)

df['features'] = df['features'].apply(clean_features)
```

```python
df['feature_count'] = df['features'].apply(
    lambda x: len(str(x).split('|')) if pd.notna(x) and x != '' else 0
)
```

```python
def clean_rating(val):
    if val == 'O': return 5.0
    if val == 'A': return 4.0
    try:
        return float(val)
    except:
        return np.nan

df['rating'] = df['rating'].apply(clean_rating)
display(df.head())
```

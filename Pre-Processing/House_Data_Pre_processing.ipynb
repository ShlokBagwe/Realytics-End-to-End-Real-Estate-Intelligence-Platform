```python
def priceCleaning(text):

    text = text.replace("Rs. ", "").strip()

    if "Crore" in text:
        text = text.replace(" Crore", "").strip()
        return round(float(text),2)

    if "Lac" in text:
        text = text.replace(" Lac", "").strip()
        return round(float(text) / 100,2)


df1['price'] = df1['price'].apply(priceCleaning)
```

## area

```python
def parse_area(val):
    if pd.isna(val):
        return np.nan, np.nan
    first_part = str(val).split('/')[0].strip()  # 4800 sqft (Plot Area)
    num = float(first_part.split('sqft')[0].strip())
    if 'Plot' in val:       atype = 'plot'
    elif 'Built' in val:    atype = 'buildup'
    elif 'Carpet' in val:   atype = 'carpet'
    else:                   atype = 'unknown'
    return num, atype

df1[['area_sqft', 'area_type']] = df1['area'].apply(
    lambda x: pd.Series(parse_area(x))
)
```

```python
def parse_age_possession(val):
    if pd.isna(val):
        return np.nan, np.nan

    parts = str(val).split('/')
    age = parts[0].strip()
    possession = parts[1].strip() if len(parts) > 1 else age

    return age, possession

df1[['property_age', 'possession_status']] = df1['agePossession'].apply(
    lambda x: pd.Series(parse_age_possession(x))
)
df1.drop(columns=['agePossession'], inplace=True)
```

## furnishing

```python
def furnishDetail(text):
    if pd.isna(text): return "Unknown"

    text_lower = text.lower()

    if "unfurnished" in text_lower:
        return "Unfurnished"
    elif "semi-furnished" in text_lower:
        return "Semi Furnished"
    elif "furnished" in text_lower:
        return "Furnished"
    else:
        return "Unknown"

df1['furnishDetails'] = df1['furnishDetails'].apply(furnishDetail)
```

## rating

```python
df1['rating'].value_counts()
```

```python
rating_map = {
    'O': 5,
    'A': 4
}

def parse_rating(val):
    if pd.isna(val): return np.nan
    if val in rating_map: return rating_map[val]
    try: return float(val)
    except: return np.nan

df1['rating'] = df1['rating'].apply(parse_rating)
```

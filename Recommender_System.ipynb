```python
def filter_properties(new_df, project_name="Any", location="Any", bhk="Any", property_type="Any", furnishing="Any",
                      selected_popular_amenities=None, selected_luxury_amenities=None, selected_family_amenities=None, min_price=None, max_price=None):

    filtered_df = new_df.copy()

    filters = {
        "Project Name": project_name,
        "Area Name": location,
        "bedroom": bhk,
        "Type of Property": property_type,
        "furnished Type": furnishing
    }

    for col, value in filters.items():
        if value != "Any":
            filtered_df = filtered_df[filtered_df[col] == value]

    if min_price is not None:
      filtered_df = filtered_df[filtered_df["Price"] >= min_price]

    if max_price is not None:
      filtered_df = filtered_df[filtered_df["Price"] <= max_price]

    if selected_popular_amenities:
        for amenity in selected_popular_amenities:
            filtered_df = filtered_df[filtered_df[amenity] == 1]

    if selected_luxury_amenities:
        for amenity in selected_luxury_amenities:
            filtered_df = filtered_df[filtered_df[amenity] == 1]

    if selected_family_amenities:
        for amenity in selected_family_amenities:
            filtered_df = filtered_df[filtered_df[amenity] == 1]



    return filtered_df
```

```python
for col in new_df.columns:
    if new_df[col].dtype == 'object':
        if set(new_df[col].unique()).issubset({'Y', 'N', np.nan}):
            new_df[col] = new_df[col].map({'Y': 1, 'N': 0}).fillna(0).astype(int)

new_df.dropna(subset=['Bathroom'], inplace=True)
new_df = new_df.reset_index(drop=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
num_transformer = Pipeline([
    ('scaler', StandardScaler())
])
cat_transformer = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_cols),
        ('cat', cat_transformer, cat_cols)
    ],
    remainder='passthrough'
)
```

```python
X = preprocessor.fit_transform(new_df.drop('Project Name', axis=1))
```

```python
X.shape
```

```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(metric='cosine', n_neighbors=11)
nn.fit(X)

print(f"Model fitted successfully with matrix X shape: {X.shape}")
```

```python
def recommend_properties(property_idx, n_recommendations=10):

    distances, indices = nn.kneighbors(
        X[property_idx].reshape(1, -1),
        n_neighbors=n_recommendations + 1
    )

    recommendations = new_df.iloc[indices[0][1:]].copy()

    #  similarity score
    recommendations['Similarity (%)'] = (
        (1 - distances[0][1:]) * 100
    ).round(2)

    return recommendations[
        [
            'Project Name',
            'Area Name',
            'Price',
            'bedroom',
            'Type of Property',
            'Similarity (%)'
        ]
    ]
```

## Recommender with Manual Weights

- **NOTE: We arent assigning weights to price and we arent scaling it. This will keep Price raw and the difference between 3cr and 15cr will be clearly seen**
- Tried with scaling but it wasnt giving priority to price

```python
from sklearn.preprocessing import MinMaxScaler

COLUMN_WEIGHTS = {

    # Numerical
    'Price':        1.0, # NOTE: We arent assigning weights to price and we arent scaling it
    'bedroom':      3.0,
    'Covered Area': 3.0,
    'sqft Price':   1.0,
    'Bathroom':     2.0,
    'Carpet Area':  1.0,
    'balconies':    0.5,
    'floor_ratio':  0.3,

    # Categorical
    'Area Name':        3.0,
    'furnished Type':   2.0,
    'Type of Property': 2.0,
    'Transaction Type': 1.0,
    'Facing':           0.5,
    'Parking_Status':   0.5,

    # Binary
    'Society':          1.0,
    'Swimming Pool':    1.0,
    'Gymnasium':        1.0,
    'Club House':       0.8,
    'Park':             0.8,
    'Security':         1.0,
    'Lift':             0.8,
    'Power Back Up':    0.5,
    'Visitor Parking':  0.5,
    'Piped Gas':        0.3,
}
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, TargetEncoder


def build_sklearn_weighted_matrix(df, col_weights):
    parts = []
    scaler = MinMaxScaler()
    ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    te = TargetEncoder(cv=3)

    SKIP_SCALING = ['Price']

    # Pre-calculate area encoding if needed so it's available for the loop
    area_encoded = None
    if 'Area Name' in df.columns and 'Price' in df.columns:
        area_encoded = te.fit_transform(df[['Area Name']], df['Price'])

    for col, weight in col_weights.items():
        if col not in df.columns:
            continue

        if col == 'Area Name':
            if area_encoded is not None:
                parts.append(area_encoded * weight)

        elif df[col].dtype == 'object':
            encoded = ohe.fit_transform(df[[col]])
            parts.append(encoded * weight)

        elif col in SKIP_SCALING:
            parts.append(df[[col]].values * weight)

        else:
            scaled = scaler.fit_transform(df[[col]])
            parts.append(scaled * weight)

    return np.hstack(parts)

X_weighted_final = build_sklearn_weighted_matrix(new_df, COLUMN_WEIGHTS)
print(f"Final Weighted Matrix Shape: {X_weighted_final.shape}")

nn_model = NearestNeighbors(metric='cosine', n_neighbors=11, algorithm='brute')
nn_model.fit(X_weighted_final)
```

```python
def recommend_v2(property_idx, n=10):

    query_vector = X_weighted_final[property_idx].reshape(1, -1)


    distances, indices = nn_model.kneighbors(query_vector, n_neighbors=n+1)


    rec_indices = indices[0][1:]
    rec_distances = distances[0][1:]

    results = new_df.iloc[rec_indices].copy()

    results['Similarity (%)'] = ((1 - rec_distances) * 100).round(2)

    return results[[
        'Project Name', 'Area Name', 'Price','Property',
        'bedroom', 'Type of Property', 'furnished Type', 'Similarity (%)'
    ]]


display(recommend_v2(9761))
```

```python
def filter_properties(new_df, project_name="Any", location="Any", bhk="Any", property_type="Any", furnishing="Any", min_price=None, max_price=None):
    filtered_df = new_df.copy()

    filters = {
        "Project Name": project_name,
        "Area Name": location,
        "bedroom": bhk,
        "Type of Property": property_type,
        "furnished Type": furnishing
    }

    for col, value in filters.items():
        if value != "Any":
            filtered_df = filtered_df[filtered_df[col] == value]

    if min_price is not None:
        filtered_df = filtered_df[filtered_df["Price"] >= min_price]
    if max_price is not None:
        filtered_df = filtered_df[filtered_df["Price"] <= max_price]

    return filtered_df

# --- EASY WORKFLOW EXAMPLE ---
# Step 1: User filters properties (e.g., looking for 3 BHK in Amboli)
search_results = filter_properties(new_df, location="Amboli", bhk=3)

# Step 2: User selects one property from the results (we'll take the first one as an example)
if not search_results.empty:
    selected_property_index = search_results.index[0]
    print(f"User selected property at index: {selected_property_index}")

    # Step 3: Display recommendations based on that specific selection
    recommendations = recommend_v2(selected_property_index, n=5)

    print("\nProperties similar to your selection:")
    display(recommendations)
else:
    print("No properties found matching those filters.")
```
